Preprocessing:

I preprocessed and mined the data to get multiple different features that would allow me to make correlations. Firstly I got the change for the day which was the open price subtracted by the closing price for the day. I created a new column called change and a categorical colum called pos_neg which was 1 for + and -1 for -. In addition i added a day before column that is the change for the day before becuase we are trying to predict the change for today based on the change for the day before

```python
stocks = [apple, google, microsoft, tesla, nvidia, amazon, meta]

#for each data table create a new attribute called change
for data in stocks:
  open = data['Open']
  close = data['Close']
  #change is = the open - the close price
  data['Change'] = data['Open'] - data['Close']
  print(data['Change'])

  #make a categorical value of pos_neg that is 1 if its + or -1 if -
  data['pos_neg'] = 0
  data.loc[data['Change'] > 0, 'pos_neg'] = 1
  data.loc[data['Change'] < 0, 'pos_neg'] = -1
  data.loc[data['Change'] == 0, 'pos_neg'] = 0
  #shift the change to the day before
  data['day_before'] = data['Change'].shift(1)
```

I wanted to analyze this data first before preprocessing more so I can know where to set my sights. After notcing the correlations between them I decided to continue by creating new categories

This next step in the preprocessing process involved creating a column called vol_change which is 1 if the volume for that day is greater than the average or -1 if less than the average, and also shifted so it is for the previous day

```python
#create a new attribute vol change that is equal to 1 if the volume is greater than average or -1 if not
#shift all the values over to get the days before value
for data in stocks:
  data['Volume'].fillna(0)
  mean = data['Volume'].mean()
  data['vol_change']= mean-data['Volume']
  data.loc[data['vol_change'] > 0, 'vol_change'] = 1
  data.loc[data['vol_change'] < 0, 'vol_change'] = -1
  data['vol_change'] = data['vol_change'].shift(1)

  data.drop(index=data.index[0], inplace=True)
apple.head(50)
```
